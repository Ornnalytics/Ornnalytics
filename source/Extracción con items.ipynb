```python
from riotwatcher import LolWatcher
import cassiopeia as cass
import pandas as pd
import numpy as np
import json 
import time
import timeit
```

```python
key = ""
watcher = LolWatcher(key)
cass.set_riot_api_key(key)
```

```python
user=watcher.summoner.by_name("euw1", "BrokenBladeee")
user
```

```python
class Summoner:
    def __init__(self, name, summonerid,puuid):
        self.name=name
        self.summonerid = summonerid
        self.puuid = puuid

first_summ = Summoner(user['name'], user['id'], user['puuid'])
#first_summ = Summoner("Name", 'id', 'puuid')        
queue=[]
queue.append(first_summ)
selected_region='euw1'
count=0
f = open('data_match.json')
matchinfo = json.load(f)
f.close()
f = open('data_user.json')
userlist = json.load(f)
f.close()

if len(list(userlist.keys())) != 0:
    summoner_visited=set(list(userlist.keys()))
else:
    summoner_visited=set()
    
if len(list(matchinfo.keys())) != 0:
    match_visited=set(list(matchinfo.keys()))
else:
    match_visited=set()
    
new_match_info={}
    
while(1):
    count += 1
    print("It: ", count)
    summ = queue.pop(0)
    userlist[summ.summonerid] = (summ.puuid, watcher.league.by_summoner(selected_region,summ.summonerid)) #call
    match_history = watcher.match.matchlist_by_puuid(region=selected_region, puuid=summ.puuid ,queue=420,count=8) #call
    for match in match_history:
        new_match_info = {}
        if match not in match_visited:  
            match_info = watcher.match.by_id(selected_region, match) #call
            new_match_info['metadata'] = match_info['metadata']
            new_match_info['info'] = {}
            new_match_info['info']['matchId'] = match_info['metadata']['matchId']
            new_match_info['info']['gameDuration'] = match_info['info']['gameDuration']
            new_match_info['info']['participants'] = []
            for participant in match_info['info']['participants']:
                participant_dict = {}
                participant_dict['item0']=participant['item0']
                participant_dict['item1']=participant['item1']
                participant_dict['item2']=participant['item2']
                participant_dict['item3']=participant['item3']
                participant_dict['item4']=participant['item4']
                participant_dict['item5']=participant['item5']
                participant_dict['item6']=participant['item6']
                participant_dict['championId']=participant['championId']
                participant_dict['championName']=participant['championName']
                participant_dict['individualPosition']=participant['individualPosition']
                participant_dict['lane']=participant['lane']
                participant_dict['puuid']=participant['puuid']
                participant_dict['role']=participant['role']
                participant_dict['summonerId']=participant['summonerId']
                participant_dict['summonerName']=participant['summonerName']
                participant_dict['teamPosition']=participant['teamPosition']
                participant_dict['win']=participant['win']
                new_match_info['info']['participants'].append(participant_dict)
                
                if participant['summonerId'] not in summoner_visited: #Nos interesa tener un set mayor que una queue
                    summoner_visited.add(participant['summonerId'])
                    queue.append(Summoner(participant['summonerName'],participant['summonerId'],participant['puuid']))
    
            ###############ADD info to dict############
            if len(userlist[summ.summonerid][1]) == 0:
                matchinfo[new_match_info['metadata']['matchId']] = (new_match_info, "UNRANKED")
            elif len(userlist[summ.summonerid][1]) == 1:
                matchinfo[new_match_info['metadata']['matchId']] = (new_match_info, userlist[summ.summonerid][1][0]['tier'])
            else:
                matchinfo[new_match_info['metadata']['matchId']] = (new_match_info, userlist[summ.summonerid][1][1]['tier'])
            
    if count%10==0:
        json_object = json.dumps(matchinfo, indent = 2) 
        with open("data_match.json", "w") as outfile:
            outfile.write(json_object)
            json_object = json.dumps(userlist, indent = 2) 
        with open("data_user.json", "w") as outfile:
            outfile.write(json_object)
```
